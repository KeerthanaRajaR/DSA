# Stack implementation using linked list

```python

class Node:
    def __init__(self,val):
        self.id = val
        self.next = None    

class Stack:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def push(self,val): 
        if(self.head is None):
            self.head = Node(val)  
            self.tail = self.head  
        else:
            self.tail.next = Node(val) 
            self.tail = self.tail.next 
    
        
    def top(self):
        return self.tail.id
    
    def Print(self):
        temp = self.head  
        while(temp): 
            print(temp.id,"->") 
            temp = temp.next 
            
    def pop(self):
        if(self.head is self.tail):
            self.head = None 
            self.tail = None 
            return
        temp = self.head 
        while(temp.next.next):
            temp = temp.next 
        self.tail = temp 
        self.tail.next = None
        
    def size(self):
        count = 0
        temp = self.head
        while(temp):
            count+=1
            temp = temp.next
        return (count)
stack = Stack()
stack.push(10)
stack.push(25)
stack.push(37)
stack.push(45)
print("size of the stack is:",stack.size())
stack.Print()
print("pop One Element:")
stack.pop()
stack.Print()
print("size of the stack is:",stack.size())
print("pop One Element:")
stack.pop()
print("Top Element:",stack.top())
stack.Print()
print("pop One Element:")
stack.pop()
stack.Print()    
```

# stack implementation using Doubly linked list

```python
class Nodedll:
    def __init__(self,val):
        self.id = val
        self.prev = None
        self.next = None
class Stackdll:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def push(self,val):
        if(self.head is None):
            self.head = Nodedll(val)
            self.tail = self.head 
        else:
            temp = self.tail 
            self.tail.next = Nodedll(val) 
            self.tail = self.tail.next 
            self.tail.prev = temp 
            
    def Print(self):
        temp = self.head 
        while(temp): 
            print(temp.id) 
            temp = temp.next 
          
            
    def pop(self):
        if(self.head is self.tail):
            self.head = self.tail = None 
        else:
            self.tail = self.tail.prev 
            self.tail.next = None 
           
    def top(self):
        return self.tail.id 
    
    def size(self):
        count = 0
        temp = self.head
        while(temp):
            count+=1
            temp = temp.next
        return (count)
        
def main():
    stack = Stackdll()
    stack.push(10)
    stack.push(15)
    stack.push(20)
    stack.push(25)
    stack.Print()
    print("size of the stack is:",stack.size())
    
    print("Top element:",stack.top())
    print("After Pop:")
    stack.pop()
    stack.Print()
    print("size of the stack is:",stack.size())
    print("After Pop:")
    stack.pop()
    stack.Print()
    print("After Pop:")
    stack.pop()
    stack.Print()
    
main()           
        
        
```

# Queue Implementation using singly Linked List

```python
class Queue:
    def __init__(self):
        self.head = self.tail = None
    
    def push(self,val):
        if(self.tail is None):
            self.head = self.tail = Node(val) 
        else:
            self.tail.next = Node(val) 
            self.tail = self.tail.next 
        
    def pop(self):
        if(self.head is None):
            print("Queue is empty")
        else:
            self.head = self.head.next 
            
    def Print(self):
        temp = self.head 

        while(temp): 
            print(temp.id) 
            temp = temp.next 
            
    def size(self):
        count = 0
        temp = self.head
        while(temp):
            count+=1
            temp = temp.next
        return (count)
        
    
        
queue = Queue()
queue.push(10)
queue.push(20)
queue.push(30)
queue.push(40)
queue.Print()
queue.pop()
print("Queue after Pop:")
queue.Print() 
print("size of the Queue is:",queue.size())
queue.pop()
print("Queue after Pop:")
queue.Print() 
queue.pop()
print("Queue after Pop:")
queue.Print() 
print("size of the Queue is:",queue.size())
queue.pop()
print("Queue after Pop:")
queue.Print() 
queue.pop()
```

# Implementation of Queue using Doubly LinkedList

```python
class NodeDLL:
    def __init__(self,val):
        self.id = val
        self.prev = None
        self.next = None

class QueueDLL:
    def __init__(self):
        self.head = self.tail = None
        
    def push(self,val): 
        if(self.tail is None):
            self.tail = NodeDLL(val)
            self.head = self.tail
        else:
            temp = self.tail
            self.tail.next = NodeDLL(val)
            self.tail = self.tail.next
            self.tail.prev = temp
    
    def Print(self):
        temp = self.head
        while(temp):
            print(temp.id)
            temp = temp.next
    
    def pop(self):
        if(self.head is None):
            print("Queue is empty")
        else:
            self.head = self.head.next
            self.head.prev = None
            
    def size(self):
        count = 0
        temp = self.head
        while(temp):
            count+=1
            temp = temp.next
        return (count)


queueDLL = QueueDLL()
queueDLL.push(1)
queueDLL.push(2)
queueDLL.push(3)
queueDLL.push(4)
queueDLL.Print()
queueDLL.pop()
print("Queue after Pop:")
queueDLL.Print()
print("size of the Queue is:",queueDLL.size())
queueDLL.pop()
print("Queue after Pop:")
queueDLL.Print()
print("size of the Queue is:",queueDLL.size())
queueDLL.pop()
print("Queue after Pop:")
queueDLL.Print()
```
